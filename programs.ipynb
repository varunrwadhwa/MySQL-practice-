```python
# factorial using recursion
def factorial(n):
	
	# single line to find factorial
	return 1 if (n==1 or n==0) else n * factorial(n - 1);

# Driver Code
num = 5;
print("Factorial of",num,"is",
factorial(num))



# factorial by iteration
def factorial(n):
	if n < 0:
		return 0
	elif n == 0 or n == 1:
		return 1
	else:
		fact = 1
		while(n > 1):
			fact *= n
			n -= 1
		return fact

# Driver Code
num = 5;
print("Factorial of",num,"is",
factorial(num))


```

```python
# Armstrong number 

# A armstrong number is a number that whos sum of the cube of the indivisual digit lead to the same number 
# 153 = 1**3 + 5**3 + 3**3 = 153 + Armstrong number 

num = int(input("Enter a number: "))


sum = 0

# find the sum of the cube of each digit
temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** 3
   temp //= 10

# display the result
if num == sum:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```

```python
# python function to find maximumin arr[] of size n
def largest(arr,n):
  
    # Initialize maximum element
    max = arr[0]
  
    # Traverse array elements from second
    # and compare every element with 
    # current max
    for i in range(1, n):
        if arr[i] > max:
            max = arr[i]
    return max
  
# Driver Code
arr = [10, 324, 45, 90, 9808]
n = len(arr)
Ans = largest(arr,n)
print ("Largest in given array is",Ans)
  
```

```python
# find permuttion of words

def permute(s, answer):
	if (len(s) == 0):
		print(answer, end = " ")
		return
	
	for i in range(len(s)):
		ch = s[i]
		left_substr = s[0:i]
		right_substr = s[i + 1:]
		rest = left_substr + right_substr
		permute(rest, answer + ch)

# Driver Code
answer = ""

s = "varun"

print("All possible strings are : ")
permute(s, answer)
```

```python
#circular linkedlist , tail points head , tail is last element , 



#Represents the node of the list.    
class Node:    
    def __init__(self,data):    
        self.data = data;    
        self.next = None;    
     
class CreateList:    
    #Declaring head and tail pointer as null.    
    def __init__(self):    
        self.count = 0;    
        self.head = Node(None);    
        self.tail = Node(None);    
        self.head.next = self.tail;    
        self.tail.next = self.head;    
        
    #This function will add the new node at the end of the list.    
    def add(self,data):    
        newNode = Node(data);    
        #Checks if the list is empty.    
        if self.head.data is None:    
            #If list is empty, both head and tail would point to new node.    
            self.head = newNode;    
            self.tail = newNode;    
            newNode.next = self.head;    
        else:    
            #tail will point to new node.    
            self.tail.next = newNode;    #this is to point existing tail to new one 
            #New node will become new tail.    
            self.tail = newNode;    
            #Since, it is circular linked list tail will point to head.    
            self.tail.next = self.head;    
                
    #This function will count the nodes of circular linked list    
    def countNodes(self):    
        current = self.head;    
        self.count=self.count+1;    
        while(current.next != self.head):    
            self.count=self.count+1;    
            current = current.next;    
        print("Count of nodes present in circular linked list: "),    
        print(self.count);    
        
     
class CircularLinkedList:    
    cl = CreateList();    
    #Adds data to the list    
    cl.add(1);    
    cl.add(2);    
    cl.add(4);    
    cl.add(1);    
    cl.add(2);    
    cl.add(3);    
    #Displays all the nodes present in the list    
    cl.countNodes();
```
